```python
#Q5数据预处理 IV：为了配合旧制中考（语数英满分150分，物理满分90分，化学60分）部分考次不是以百分制计分，
#                为与之前的百分制计分保持一致，请将部分非百分制计分的列转换为百分制计分。
#提示1：可以通过查看每列的最大值查找非百分制计分的数据列。
#提示2：8年级第2学期的期末考试中，语数英为120分制。

data_4=data_3.copy()
for index in ('CHI822', 'MATH822', 'ENG822'):
    data_4[index]=data_4[index].astype('float')
    i=0
    for ele in data_4[index]:
        ele=round(ele*100/120)
        data_4.loc[i, index]=ele
        i+=1

for index in ('CHI611', 'MATH611', 'ENG611', 'CHI612', 'MATH612', 'ENG612','CHI621', 'MATH621', 'ENG621', 'CHI622', 'MATH622', 'ENG622',
                     'CHI711', 'MATH711', 'ENG711', 'CHI712', 'MATH712', 'ENG712','CHI721', 'MATH721', 'ENG721', 'CHI722', 'MATH722', 'ENG722',
                     'CHI811', 'MATH811', 'ENG811', 'PHY811','CHI812', 'MATH812', 'ENG812', 'PHY812','CHI821', 'MATH821', 'ENG821', 'PHY821','PHY822', 
                     'CHI911', 'MATH911', 'ENG911', 'PHY911', 'CHE911','CHI912', 'MATH912', 'ENG912', 'PHY912', 'CHE912','CHI921', 'MATH921', 'ENG921', 'PHY921','CHE921'):
    data_4[index]=data_4[index].astype('float')
    F=data_4[index].max()
    
    if F<=90.0 or F>100.0:
        i=0
        for ele in data_4[index]:
            if F>100:
                ele=round(ele*100/150)
            elif F<=60: 
                ele=round(ele*100/60)
            else:
                ele=round(ele*100/90)
            data_4.loc[i, index]=ele
            i+=1
data_4.to_csv('Q5.csv', index=0)
```
